```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", 
                                on=["school_name", "school_name"])
school_data_complete.head()
```

```python
#Calculate the total number of schools
AllSchool = school_data_complete["school_name"].unique() 
Total_schools = len(AllSchool)
```

```python
#Calculate the total number of students
Students = len(school_data_complete["student_name"])
Total_students = "{:,}".format(Students)
```

```python
#Calculate the total budget
Total_budget = "${:,.2f}".format(school_data_complete["budget"].sum())
```

```python
#Calculate the average math score
Total_math = school_data_complete["math_score"].mean()
Average_math = round(Total_math,6)
```

```python
#Calculate the average reading score
Total_read = school_data_complete["reading_score"].mean()
Average_read = round(Total_read,6)
```

```python
#Calculate the overall passing rate (overall average score),
#i.e. (avg. math score + avg. reading score)/2
Average_total = round((Average_math+Average_read)/2,6)
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
Passing_math = school_data_complete.loc[school_data_complete["math_score"] >= 70]
Percentage_math = round((len(Passing_math)/Students)*100,6)
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
Passing_read = school_data_complete.loc[school_data_complete["reading_score"] >= 70]
Percentage_read = round((len(Passing_read)/Students)*100,6)
```

```python
#Create a dataframe to hold the above results

Result = [{"Total Schools": Total_schools,
           "Total Students":Total_students,
           "Total Budget":Total_budget,
           "Average Math Score":Average_math,
           "Average Reading Score":Average_read,
           "% Passing Math":Percentage_math,
           "% Passing Reading":Percentage_read,
           "% Overall Passing Rate":Average_total}]
Result_df =pd.DataFrame(Result)

#Optional: give the displayed data cleaner formatting
Result_df = Result_df [["Total Schools","Total Students",
              "Total Budget","Average Math Score",
              "Average Reading Score",
              "% Passing Math","% Passing Reading",
              "% Overall Passing Rate"]]
Result_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
##Perform calculation of math, reading and overall scores based on different criteria 
def math_read_overall_df(perform_df):
    Average_math_value = perform_df["Average Math Score"].mean()
    Average_read_value = perform_df["Average Reading Score"].mean()
    Passing_math_value = perform_df["% Passing Math"].mean()
    Passing_read_value = perform_df["% Passing Reading"].mean()
    Passing_overall_value = perform_df["% Overall Passing Rate"].mean()

    summary_df = pd.DataFrame({ "Average Math Score": Average_math_value,
                                "Average Reading Score": Average_read_value,
                                "% Passing Math": Passing_math_value,
                                "% Passing Reading": Passing_read_value,
                                "% Overall Passing Rate": Passing_overall_value})
    return (summary_df)
```

```python
# Sample bins. Feel free to create your own bins.
bins = [0, 585, 615, 645, 675]
# Create the names for the  bins
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
#Find the lower limit for the bin
Each_school_summary["Per Student Budget"].min()
```

```python
#Find the upper limt range for the bin
Each_school_summary["Per Student Budget"].max()
```

```python
Each_school_summary["Spending Ranges (Per Student)"] = pd.cut(per_student_budget, bins, labels=group_names)
```

```python
#ACreate a group based on average Spending Ranges (Per Student)
Spending_range_gb = Each_school_summary.groupby(["Spending Ranges (Per Student)"])
```

```python
# create DataFrame -  based on average Spending Ranges (Per Student) and display the result
#Call 'math_read_overall_df' function from line 28
Average_spending_summary_df = math_read_overall_df(Spending_range_gb)
Average_spending_summary_df.reset_index(inplace=True)
Average_spending_summary_df
```
